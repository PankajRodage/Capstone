# Detailed Tasks:

1.	Understanding Order Amount Distribution:
○	Task: Load the dataset and extract the order amount feature. Use histograms to visualize the distribution of order amounts. Create box plots to identify outliers.
- ○	Steps:
 - ■	Load the dataset using pandas.
 - ■	Extract the order amount column.
 - ■	Use matplotlib to create a histogram of order amounts.
 - ■	Use seaborn to create a box plot to identify outliers.


```python
# Load the dataset using pandas
df = pd.read_csv('order_items_dataset.csv')
```

```python
df
```

```python
# Extract the order amount column
df['price']
```

```python
# Use matplotlib to create a histogram of order amounts.
plt.figure(figsize=(6,3))
plt.hist(data=df, x='price',bins=1000,color='b',edgecolor='r')
plt.xlabel("Order Amount OR Price")
plt.title("Order Amount Distribution OR Price Distribution")
plt.xlim(0,3000)
plt.show()
```

```python
# Use seaborn to create a box plot to identify outliers.
plt.figure(figsize=(6,3))
sns.boxplot(df.price,color='pink')
plt.show()
```

```python
Q1=df['price'].quantile(0.25)
Q3=df['price'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1-1.5*IQR
upper_bound = Q3+1.5*IQR

df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
df
df.price.describe()
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(df.price,color='pink')
plt.show()
```

```python
plt.figure(figsize=(14,10))
plt.hist(data=df, x='price',bins=40,color='b',edgecolor='r')
plt.xlabel("Order Amount OR Price")
plt.title("Order Amount Distribution OR Price Distribution")
plt.xlim(0,400)
plt.ylim(0,7000)

plt.show()
```

# 4.	Statistical Summary of Delivery Times:
○	Task: Calculate basic statistics (mean, median, mode, standard deviation) for delivery times and visualize the distribution using histograms and box plots.
- ○	Steps:
- ■	Load the dataset using pandas.
- ■	Extract the delivery time column.
- ■	Calculate mean, median, mode, and standard deviation.
- ■	Use matplotlib to create a histogram of delivery times.
- ■	Use seaborn to create a box plot of delivery times.


```python
# Load the dataset using pandas.
df3 = pd.read_csv('orders_dataset.csv')
df3
```

```python
# Extract the delivery time column.
df3['order_purchase_timestamp'] = pd.to_datetime(df3['order_purchase_timestamp'])
df3['order_delivered_customer_date'] = pd.to_datetime(df3['order_delivered_customer_date'])
df3['order_purchase_timestamp']
df3['order_delivered_customer_date']
```

```python
df3['delivery_days'] = (df3['order_delivered_customer_date'] - df3['order_purchase_timestamp']).dt.days
df3['delivery_days']
```

```python
# Calculate mean, median, mode, and standard deviation.
df3['delivery_days'].mean() # Mean
```

```python
df3['delivery_days'].median() # Median
```

```python
df3['delivery_days'].mode() # Mode
```

```python
df3['delivery_days'].std() #
```

```python
# Use matplotlib to create a histogram of delivery times.
plt.figure(figsize=(14,10))
plt.hist(data=df3, x='delivery_days',bins=50,color='b',edgecolor='r')
plt.xlabel("Delivery Days")
plt.title("Delivery Days Distribution")
plt.show()
```

```python
# Use seaborn to create a box plot of delivery times.
plt.figure(figsize=(14,10))
sns.boxplot(df3.delivery_days, color='pink')
plt.show()
```

```python
# calculation to remove outliers from 'delivery_days' column and plot the box plot after removing the outliers
df_time_filtered = df3[(df3['delivery_days'] >= 0) & (df3['delivery_days'] <= 50)]
plt.figure(figsize=(14,10))
df_time_box = sns.boxplot(df_time_filtered['delivery_days'], color='pink')
plt.show()
```

# 8.	Order Value by Product Category:
○	Task: Analyze how order values vary across different product categories using box plots and bar charts.
- steps:
 - Load the dataset using pandas.
 - Extract order value and product category columns.
 - Use seaborn to create box plots of order value by product category.
 - Use matplotlib to create bar charts of average order value by product category.


```python
# Merge two data frames
df6 = pd.merge( df4,df1, on='product_id', how='inner')
df6
```

```python
# Extract order value.
order_value  = df6['price']
order_value
```

```python
# Extract product category columns.
product_category = df6['product_category_name']
product_category
```

```python
# Use seaborn to create box plots of order value by product category.
plt.figure(figsize=(20,18))
sns.boxplot(x=product_category, y=order_value)
plt.title('Order Value by Product Category (Box Plot)')
plt.xlabel('Product Category')
plt.ylabel('Order Value')
plt.xticks(rotation=90)
plt.show()
```

```python
# Use matplotlib to create bar charts of average order value by product category.
order_value_avg = df6.groupby('product_category_name')['price'].mean()
plt.figure(figsize=(14,10))
order_value_avg.plot(kind='bar')
plt.title('Average Order Value by Product Category (Bar Chart)')
plt.xlabel('Product Category')
plt.ylabel('Average Order Value')
plt.show()
```

# 9.	Customer Segments and Product Categories:
○	Task: Examine the relationship between customer segments and product categories using cross-tabulations and heatmaps.
- Steps:
- Load the dataset using pandas.
- Extract customer segment and product category columns.
- Create cross-tabulations using pandas.
- Use seaborn to create heatmaps of the cross-tabulations.


```python
# Load the dataset using pandas.
df7 = pd.read_csv('customers_dataset.csv')
df7
```

```python
# merge customer dataset and order dataset
df8 = pd.merge( df7,df3, on='customer_id', how='inner')
df8
```

```python
# merged customer and order datasets are merge into order_items dataset
df9 = pd.merge( df8,df4, on='order_id', how='inner')
df9
```

```python
# merged customer,order,order_items dataset into product dataset
df10 = pd.merge( df9,df1, on='product_id', how='inner')
df10
```

```python
# Extract customer state 
customer_state = df10['customer_state']
customer_state
```

```python
# Extract product category columns.
product_category = df10['product_category_name']
product_category
```

```python
# Create cross-tabulations using pandas.
cross_tab = pd.crosstab(customer_state, product_category)
cross_tab
```

```python
# Use seaborn to create heatmaps of the cross-tabulations.
plt.figure(figsize=(30,8))
sns.heatmap(cross_tab,annot=True,cmap='coolwarm', square=True)
plt.title('Customer States vs. Product Categories')
plt.xlabel('Product Categories')
plt.ylabel('Customer States')
plt.show()
```
